# **Generación de tokens de palabras**

```
from keras.preprocessing.text import Tokenizer
# Es sensible a los acentos

samples = ['El aprendizaje máquina es un área de la inteligencia artificial', 'El aprendizaje profundo es un área del aprendizaje máquina']
 
# Creamos un tokenizer, a partir del cual cada palabra de una texto, se convertirá en un valor númerico (index)
tokenizer = Tokenizer(num_words=1000)   # Se define con cuantas palabras vamos a estar trabajando (1000)
# Construimos el diccionario que permite obtener el valor número (index) de una palabra
tokenizer.fit_on_texts(samples)

# Las listas de strings se convierten en listas de índices
sequences = tokenizer.texts_to_sequences(samples)   

word_index = tokenizer.word_index          # Mostramos el diccionario de palabra-index
print('Se encontrarion %s tokens únicos.' % len(word_index))
print(word_index)
```
